## Data Exploration

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df['Weight'].fillna(df['Weight'].mean, inplace=True)
```

## Creating a new feature

```python
top_ten = df['NOC'].value_counts().head(10)
top_ten
```

## Participation by National Olympic Committees

```python
nocdf = pd.DataFrame(top_ten)
nocdf
```

## Visualizing the data

```python
sns.barplot(data=nocdf,x='NOC', hue='NOC', y='count',palette='coolwarm', legend=False)
```

## Most Medals

```python
g = df.groupby(['NOC'])['Medal'].value_counts().reset_index()
```

```python
g
```

```python
top_ten_medals = g.groupby(['NOC'])['count'].sum().reset_index().sort_values('count',ascending=False).head(10)
top_ten_medals
```

```python
plt.figure(figsize = (15, 8))
ax = sns.barplot(data = top_ten_medals, y = 'NOC', x = 'count', palette = 'flare', hue='count',legend=False)
```

```python
import warnings
warnings.filterwarnings('ignore')

summer_olympics = df[df['Season'] == 'Summer']
winter_olympics = df[df['Season'] == 'Winter']

summer_sports = summer_olympics[['Year', 'Sport']].drop_duplicates()['Sport'].value_counts()
winter_sports = winter_olympics[['Year', 'Sport']].drop_duplicates()['Sport'].value_counts()
fig, ax = plt.subplots(2, 1, figsize = (22, 10))
sns.barplot(x = summer_sports.keys(), y = summer_sports.values, 
            ax = ax[0], palette='Reds_r', hue=summer_sports.keys())
sns.barplot(x = winter_sports.keys(), y = winter_sports.values,
            ax = ax[1], palette='Blues_r', hue=winter_sports.keys())

ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation = 45)
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation = 45)
        
plt.tight_layout()
```
